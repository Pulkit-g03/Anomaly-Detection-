```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Load the dataset

data = df

# Select numerical columns for anomaly detection
numerical_features1 = ['Pid', 'Tid']
X = data[numerical_features1]
X.fillna(X.mean(), inplace=True)
# Standardize the data
scaler1 = StandardScaler()
X_scaled = scaler1.fit_transform(X)

# Apply Isolation Forest
iso_forest1 = IsolationForest(random_state=42, contamination=0.4)
data['anomaly'] = iso_forest1.fit_predict(X_scaled)

# Count anomalies (-1 indicates anomaly, 1 indicates normal)
anomaly_count = data['anomaly'].value_counts()

print("Anomaly Detection Results:")
print(anomaly_count)
```

```python
from sklearn.metrics import f1_score, r2_score, accuracy_score
from sklearn.linear_model import LinearRegression
import numpy as np

# Assuming `data['anomaly']` from Model B is the pseudo-ground truth
# Model A predictions
y_pred = df['Anomaly']  # Predictions from Model A
y_true = data['anomaly']  # Model B predictions as pseudo-ground truth

# Convert labels to binary (1 for anomaly, 0 for normal)
y_pred_binary = np.where(y_pred == -1, 1, 0)
y_true_binary = np.where(y_true == -1, 1, 0)

# Compute F1-score
f1 = f1_score(y_true_binary, y_pred_binary)
print("Estimated F1 Score:", f1)

# Compute Accuracy
accuracy = accuracy_score(y_true_binary, y_pred_binary)
print("Estimated Accuracy:", accuracy)

# R² Score using Reconstruction Error
# Assuming 'X_scaled' is the original feature set used in Model B
regressor = LinearRegression()
regressor.fit(X_scaled, y_pred_binary)  # Predicting anomalies
y_pred_reconstructed = regressor.predict(X_scaled)

# Compute R² score
r2 = r2_score(y_pred_binary, y_pred_reconstructed)
print("R² Score (Reconstruction Error):", r2)
```
